```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter product name:"))
opr = int(input("Enter original price:"))
dpr = int(input("Enter discount price:"))
ms = int(input("Enter money saved:"))
qn = int(input("Enter quantity:"))

record[prod_id] = {'product_name': name, 'original_price': opr, 'discount_price': dpr, 'money_saved': ms, 'quantity': qn}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
