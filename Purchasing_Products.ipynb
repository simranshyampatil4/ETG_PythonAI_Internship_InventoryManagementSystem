```python
#Read Inventory
```

```python
import json

fd = open("record.json",'r')
r = fd.read()
fd.close()

records = json.loads(r)
```

```python
records
```

```python
#Purchase
```

```python
customer_name = str(input("Name: "))
customer_ph = int(input("Mobile No.: "))
print("****************Welcome " + customer_name + " " + "To Our Shop****************" )
print("--------------------------------------------------------------")
ui_prod  = str(input("Enter the product_Id: "))
ui_quant = int(input("Enter the quantity: "))
for i in records.keys():
    while ui_prod == i:
        if ui_quant <= records[i]['quantity']:
            print("Product Name: ", records[ui_prod]['product_name'])
            print("Price Per Quantity: ", records[ui_prod]['discount_price'])
            print("************************************************************")
            print(" Total Billing Amount: ", records[ui_prod]['discount_price'] * ui_quant)
            print("************************************************************")
            print("Thank You " + customer_name + " " +"For Visitng Our Shop." +"\n" + "Happy Shoping!!...")
            
        else:
            print("Sorry Product Is Out of Stock " + "\n" + "Sorry for inconvenience")
        break
if ui_prod not in records.keys():
   print("No Such Product Id exists:  ",ui_prod,"\nPlease enter a valid product ID")


records[ui_prod]['quantity'] = records[ui_prod]['quantity'] - ui_quant
```

```python
#Update Inventory after purchase
```

```python
js = json.dumps(records)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

```python
#Sales JSON
```

```python
fd = open("sales.json",'r')
ts = fd.read()
fd.close()

sales = json.loads(ts)
num = len(sales)+1

sales[num] = {'prod':ui_prod, 'quanity' : ui_quant, 'amount': records[ui_prod]['discount_price'] * ui_quant}
sale = json.dumps(sales)

fd = open("sales.json",'w')
fd.write(sale)
fd.close()
```
